```python
def Positions(filename):
    list_x = []
    list_y = []
    list_z = []
    print(filename)
    file = pd.read_csv(filename)
    for i in range(0, len(file.columns)):
        if file.columns[i].find("positionX") > -1:
            list_x.append(file[file.columns[i]])
        
        if file.columns[i].find("positionY") > -1:
            list_y.append(file[file.columns[i]])
        
        if file.columns[i].find("positionZ") > -1:
            list_z.append(file[file.columns[i]])
    
    return list_x, list_y, list_z
```

```python
from scipy.interpolate import UnivariateSpline
#import numpy as np

def curvature_splines(x, y=None, error=0.1):
    """Calculate the signed curvature of a 2D curve at each point
    using interpolating splines.
    Parameters
    ----------
    x,y: numpy.array(dtype=float) shape (n_points, )
         or
         y=None and
         x is a numpy.array(dtype=complex) shape (n_points, )
         In the second case the curve is represented as a np.array
         of complex numbers.
    error : float
        The admisible error when interpolating the splines
    Returns
    -------
    curvature: numpy.array shape (n_points, )
    Note: This is 2-3x slower (1.8 ms for 2000 points) than `curvature_gradient`
    but more accurate, especially at the borders.
    """

    # handle list of complex case
    if y is None:
        x, y = x.real, x.imag

    t = np.arange(x.shape[0])
    std = error * np.ones_like(x)

    fx = UnivariateSpline(t, x, k=4, w=1 / np.sqrt(std))
    fy = UnivariateSpline(t, y, k=4, w=1 / np.sqrt(std))

    xˈ = fx.derivative(1)(t)
    xˈˈ = fx.derivative(2)(t)
    yˈ = fy.derivative(1)(t)
    yˈˈ = fy.derivative(2)(t)
    curvature = (xˈ* yˈˈ - yˈ* xˈˈ) / np.power(xˈ** 2 + yˈ** 2, 3 / 2)
    return curvature
```

```python
def Gussian(OneJoinOfaSample, outX_Y_Z, sstd, Neighbours):
    EverypointFrameNeiboursGuassian14_2std = []
    for i in range(0, len(Neighbours)):
        fractionf = 0
        fractiond = 0
        for j in range(0, len(Neighbours[i])):
            deltay = pow(OneJoinOfaSample[j] - OneJoinOfaSample[i], 2)
            deltax = pow(j - i, 2)
            face = outX_Y_Z[j]*ma.exp(-1*(deltax + deltay)/2*pow(sstd,2))
            if face == 0:
                face = outX_Y_Z[j]*np.finfo(np.float32).eps
            down = ma.exp(-1*(deltax + deltay)/2*pow(sstd,2))
            if down == 0:
                down = np.finfo(np.float32).eps

            fractionf = face + fractionf
            fractiond = down + fractiond
        EverypointFrameNeiboursGuassian14_2std.append(fractionf/fractiond)
    return np.asarray(EverypointFrameNeiboursGuassian14_2std)
```

```python
def NeiboursOfAPoint(sstd, JointNinDifferentFramesFirstSample):
    NeighboursOfaPoint = []
    for i in range(0, len(JointNinDifferentFramesFirstSample)):
        tmp = []
        for j in range(0, len(JointNinDifferentFramesFirstSample)):
            deltay = pow(JointNinDifferentFramesFirstSample[j] - JointNinDifferentFramesFirstSample[i], 2)
            deltax = pow(j - i, 2)
            f = ma.sqrt(deltax + deltay)
            if f < sstd:
                tmp.append(j)

        NeighboursOfaPoint.append(tmp)
    return NeighboursOfaPoint
```
