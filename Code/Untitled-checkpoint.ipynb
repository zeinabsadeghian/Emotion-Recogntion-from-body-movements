```python
def Positions(filename):
    list_x = []
    list_y = []
    list_z = []
    print(filename)
    file = pd.read_csv(filename)
    for i in range(0, len(file.columns)):
        if file.columns[i].find("positionX") > -1:
            list_x.append(file[file.columns[i]])
        
        if file.columns[i].find("positionY") > -1:
            list_y.append(file[file.columns[i]])
        
        if file.columns[i].find("positionZ") > -1:
            list_z.append(file[file.columns[i]])
    
    return list_x, list_y, list_z
```
